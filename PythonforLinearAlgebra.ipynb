1) come up with 3 2D vectors, tow of them are orthogonal, but neither is orthogonal to the third
2) plot all three vectors on a 2D plane - make sure the axes are square and equal

```python
vecA = np.array([2,4])
vecB = np.array([-2,1])
print(np.dot(vecA, vecB))

vecC = np.array([5,6])
print(np.dot(vecC, vecB))
print(np.dot(vecA, vecC))

plt.plot([0, vecA[0]], [0, vecA[1]], 'r-', label='vecA')
plt.plot([0, vecB[0]], [0, vecB[1]], 'g-', label='vecB')
plt.plot([0, vecC[0]], [0, vecC[1]], 'b-', label='vecC')

plt.legend()
plt.grid(True)
plt.axis('equal')   # to make sure axes are square and equal
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

## Matrix Multiplication

[MxN][NxK]=[MxK] <- these matrices can be multiplied because of N.

If their inner dimensions would not match, we cannot multiply those matrices

```python
M1 = np.random.randn(4,5)   # 4x5
M2 = np.random.randn(4,5)   # 4x5
# cannot be multiplied...
# unless we Transpose one of the matrices

print(np.matmul(M1.T,M2)) # matmul to multiply matrices
print(' ')
print(M1.T@M2)  # other way of multiplying matrices
print(' ')
print(np.dot(M1.T, M2)) # third way of multiplying matrices
```

1) Create a 3x3 matrix of integers
2) Multiply that matrix by identity matrix and then by zeros matrix and finally multiply it by this matrix transpose.

```python
M = np.random.randint(0, 10, size=(3,3))
print("M:\n", M)

M_identity = np.identity(3)
print("\nIdentity Matrix 3x3:\n", M_identity)

M_zeros = np.zeros((3,3))
print("\nZeros Matrix 3x3:\n", M_zeros)

print("\nMxM_identity:\n", M@M_identity)
print("\nMxM_zeros:\n", M@M_zeros)
print("\nMxMT:\n", M@M.T)
```

## Matrix Inverse

```python
A = np.random.randn(4,4)
Ainv = np.linalg.inv(A) # function to calculate the inverse of a given matrix

AinvA = A@Ainv

print(A)
print(' ')
print(Ainv)
print(' ')
print(AinvA)
```

```python
# Since it's hard to see it on numbers, let's visualize it on plot

fig,ax = plt.subplots(1,3,figsize=(6,5))

ax[0].imshow(A)
ax[0].set_title('A')
ax[1].imshow(Ainv)
ax[1].set_title('A$^{-1}$')
ax[2].imshow(AinvA)
ax[2].set_title('A$^{-1}$A')
```

As we can see, we obtained Identity Matrix so indeed Ainv is the inverse matrix of matrix A
